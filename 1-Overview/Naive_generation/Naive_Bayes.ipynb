# Read text with labels

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
data = df.iloc[:,0:2]
```

# Vectorize SMS into count vectors

```python
vectorizer = CountVectorizer()
sms_vectorized = vectorizer.fit_transform(data.v2)
```

# Further preparation

```python
labels = data.v1.to_numpy()
words = vectorizer.get_feature_names()
one_hots = sms_vectorized.toarray()
```

```python
spam_oh = one_hots[labels == 'spam']
ham_oh = one_hots[labels == 'ham']
```

# Count probabilities
Here we calculate probabilities $\mathbb{P} (word|target) $

```python
spam_probs = spam_oh.sum(0) / spam_oh.sum()
ham_probs = ham_oh.sum(0) / ham_oh.sum()
```

# Generate spam or ham SMS

```python
def generate_sms(target, sms_len):
    if target == 'spam':
        probs = spam_probs
    elif target == 'ham':
        probs = ham_probs
    else:
        raise ValueError('Target must be either \'spam\' or \'ham\'.')
    
    generated_sms = ''
    for i in range(sms_len):
        generated_sms += np.random.choice(words, p=probs) + ' '
        
    return generated_sms[:-1]
```

```python
generate_sms('ham', 10)
```
